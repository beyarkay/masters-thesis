```python
def ReLU(x):
    return max(x, 0.0)
    
x = np.linspace(-10, 10, 101)
y = np.vectorize(ReLU)(x)

plt.plot(x, y)
plt.title(r'$\mathtt{ReLU}(x) = \max(x, 0)$')
plt.xlabel("Function Input")
plt.ylabel("Function Output")
```
